```python
import random
import time

def simulacion_robot_inteligente():
    # Estado inicial del robot (Variables de juego)
    bateria = 100
    bolsas_caca = 5
    felicidad_perro = 0
    tiempo_paseo = 0
    meta_tiempo = 30 # Minutos objetivo

    print("--- 🤖 INICIANDO ROBOT PASEADOR 2.0 ---")

    while tiempo_paseo < meta_tiempo and bateria > 0:
        print(f"\n[Reloj: {tiempo_paseo}min | Batería: {bateria}% | Bolsas: {bolsas_caca}]")

        # Simulación de sensor: ¿Qué hay en el camino?
        suceso = random.choice(["nada", "charco", "necesidad_perro", "lluvia_ligera"])

        if suceso == "charco":
            print("❗ Obstáculo: ¡Charco detectado! Calculando ruta alternativa...")
            bateria -= 10 # Esquivar gasta más energía

        elif suceso == "necesidad_perro":
            if bolsas_caca > 0:
                print("💩 El perro hizo sus necesidades. Usando bolsa y limpiando...")
                bolsas_caca -= 1
                felicidad_perro += 20
            else:
                print("❌ ¡OH NO! No quedan bolsas. Penalización de limpieza.")
                felicidad_perro -= 10

        elif suceso == "lluvia_ligera":
            print("🌧️ Empieza a chispear. Aumentando velocidad de retorno.")
            bateria -= 5

        else:
            print("🌳 Paseo tranquilo por el parque...")
            felicidad_perro += 5

        # El tiempo avanza y la batería baja por el simple hecho de estar encendido
        tiempo_paseo += 5
        bateria -= 5
        time.sleep(1) # Para que la simulación no sea instantánea

    # --- MÉTRICAS DE EVALUACIÓN FINAL ---
    print("\n" + "="*30)
    print("RESUMEN DEL PASEO")
    print(f"Distancia/Tiempo total: {tiempo_paseo} min")
    print(f"Felicidad del perro: {felicidad_perro} pts")
    print(f"Batería restante: {bateria}%")

    if bateria <= 0:
        print("RESULTADO: 🔴 El robot se quedó sin batería en el parque.")
    else:
        print("RESULTADO: 🟢 Robot regresó a casa con éxito.")

simulacion_robot_inteligente()
```
